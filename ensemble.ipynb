```python
import json

with open('predictions_ml_ori.json') as f:
    ori = json.load(f)

with open('predictions_ml_ensemble.json') as f:
    ens = json.load(f)
    
```

```python
import pandas as pd

df_ori = pd.DataFrame(ori)
mappig ={0: 'negatif', 1: 'netral', 2: 'positif'}
for col in df_ori.columns:
    df_ori[col] = df_ori[col].map(mappig)
```

```python
df_ens = pd.DataFrame(ens)

mappig ={0: 'negatif', 1: 'netral', 2: 'positif'}
for col in df_ens.columns:
    df_ens[col] = df_ens[col].map(mappig)
```

```python
df_ens = pd.DataFrame(ori)
mappig ={0: 'negatif', 1: 'netral', 2: 'positif'}
for col in df_ori.columns:
    df_ens[col] = df_ori[col].map(mappig)
```

```python
df_ens['ensemble']=df_ens[['xgboost', 'logreg','lightgbm']].mode(axis=1)[0]      
```

```python
with open('y_test.json') as f:
    y_test = json.load(f)
y_test=pd.Series(y_test).map(mappig)
```

```python
with open('y_test_ensemble.json') as f:
    y_test = json.load(f)
y_test=pd.Series(y_test).map(mappig)
```

```python
y_test
```

```python
# Accuracy, F1, Precision, Recall
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
accuray=accuracy_score(y_test, df_ori['ensemble'])  
f1=f1_score(y_test, df_ori['ensemble'], average='weighted')
precision=precision_score(y_test, df_ori['ensemble'], average='weighted')
recall=recall_score(y_test, df_ori['ensemble'], average='macro')
print('Accuracy:', accuray, 'F1:', f1, 'Precision:', precision, 'Recall:', recall)  
```

```python
# Accuracy, F1, Precision, Recall
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
accuray=accuracy_score(y_test, df_ens['ensemble'])  
f1=f1_score(y_test, df_ens['ensemble'], average='weighted')
precision=precision_score(y_test, df_ens['ensemble'], average='weighted')
recall=recall_score(y_test, df_ens['ensemble'], average='weighted')
print('Accuracy:', accuray, 'F1:', f1, 'Precision:', precision, 'Recall:', recall)  
```
